$$\frac{\partial u}{\partial t} + u\frac{\partial u}{\partial x} = (\frac{0.01}{\pi})\frac{\partial^2 u}{\partial x^2}$$

```python
#burger equation
# u = -sin pi x
x = sn.Variable('x')
t = sn.Variable('t')
u = sn.Functional('u', [x,t], 8*[20], 'tanh')# 8 hidden layers with 20 neural units
```

$$\frac{\partial u}{\partial t} + u\frac{\partial u}{\partial x} - (\frac{0.01}{\pi})\frac{\partial^2 u}{\partial x^2} = 0$$

```python
# the optimization model
from sciann.utils.math import diff, sign, sin
```

```python
L1 = diff(u, t)+ u*diff(u, x) - (0.01/pi)*diff(u, x, order=2)
```

```python
zero = 0.001
c1 = (1+ sign(zero-t))*(u+ sin(pi*x))
c2 = (1+ sign(-1-x))*u
c3 = (1+sign(x-1))*u
```

```python
model = sn.SciModel([x,t],[L1, c1,c2,c3])
```

```python
xdata, tdata = meshgrid(linspace(-1,1,100), linspace(0,1,100))
```

```python
mtrain = model.train([xdata, tdata], 4*['zero'], learning_rate =0.002, epochs=500)
```

# PINN = Physics Informed Neural Network / deep learning

$$\frac{\partial^2 f}{\partial x^2} + \frac{\partial^2 f}{\partial y^2} + 2f(x,y) = 0$$

where
$$f(x,y) = \sin{x}\sin{y}$$

```python
from sciann import math
```

```python
fxx = math.diff(f, x, order=2)
fxx
```

```python
fyy = math.diff(f, y, order=2)
L = fxx+fyy+2*f
L
```

```python
data1 = Data(f)
data2 = Data(L)
m2 = SciModel([x,y],[data1,data2])
mt2 = m2.train([xdata, ydata],[fdata,'zero'], epochs = 100)
#mt2 = model train 2
```
